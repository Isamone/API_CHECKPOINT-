```python



##  EXTRACTING RELEVANT DATA FROM NeoWs

url2 = "https://api.nasa.gov/neo/rest/v1/neo/browse"

response2 = requests.get(url2, params=params)

data2 = response2.json()

data2

# data2["near_earth_objects"][0]["name"]

# for asteroid in data2["near_earth_objects"]:
#     print(asteroid["name"])


data2

asteroid_list = []

for asteroid in data2["near_earth_objects"]:


    if asteroid["close_approach_data"]:
        asteroid_details ={
            "Asteroid ID": asteroid["id"],
            "Asteroid name": asteroid["name"],
            "Estimated Diameter": asteroid["estimated_diameter"]["kilometers"]["estimated_diameter_min"],
            "Absolute Magnitude": asteroid["absolute_magnitude_h"],
            "Relative Velocity": asteroid["close_approach_data"][0]["relative_velocity"]["kilometers_per_second"]

        }

        #Append eac details to the main list
        asteroid_list.append(asteroid_details)

    else:
        asteroid_details ={
            "Asteroid ID": asteroid["id"],
            "Asteroid name": asteroid["name"],
            "Estimated Diameter": asteroid["estimated_diameter"]["kilometers"]["estimated_diameter_min"],
            "Absolute Magnitude": asteroid["absolute_magnitude_h"],
            "Relative Velocity": ""
        }



asteroid_list

df = pd.DataFrame(asteroid_list)

df
```
